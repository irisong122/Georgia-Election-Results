```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
def scrape(link): 
    driver = webdriver.Firefox()
    driver.get(link)

    # view by locality
    wrapper = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, "app-ballot-item-wrapper")))
    a = wrapper.find_element(By.TAG_NAME, "a")
    a.click()

    # scroll down until end of page
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(0.5)

        try:
            last_county = driver.find_element(By.LINK_TEXT, "Worth County")
        except:
            continue
        else:
            print("Last county found")
            break

    # click button to get vote counts by method
    button_wrapper = driver.find_element(By.TAG_NAME, "app-vote-method")
    button = button_wrapper.find_element(By.TAG_NAME, "button")
    button.click()

    # election year
    header = driver.find_element(By.CLASS_NAME, "election-header")
    election_name = header.find_element(By.CLASS_NAME, "h6").text

    counties = driver.find_elements(By.TAG_NAME, "app-ballot-item-wrapper")

    county_names = []
    candidate_names = []
    votes_by_category = []

    for county in counties:
        county_names.append(county.find_element(By.TAG_NAME, "a").text)
        candidate_list = county.find_elements(By.CLASS_NAME, "candidate")
        td_list = (county.find_elements(By.TAG_NAME, "td"))

        for name in candidate_list: 
            candidate_names.append(name.text)

        for td in td_list:
            try:
                votes_by_category.append(td.find_element(By.TAG_NAME, "span").text)
            except:
                continue
            
    driver.close()

    if (len(county_names) == 159):
        print("All counties found")
    else:
        print((159 - len(county_names)) + " counties found")

    # data cleaning
    election_day = []
    absentee_mail = []
    early_in_person = []
    provisional = []
    total = []

    votes = [election_day, absentee_mail, early_in_person, provisional, total]

    for i in range(len(votes_by_category)):
        votes[i % 5].append(votes_by_category[i])

    if (len(candidate_names) != len(election_day)):
        del candidate_names[0:len(set(candidate_names))]

    print(candidate_names)

    print(len(set(candidate_names)))
    print("candidates")

    county_names_full = []
    for county in county_names:
        for i in range(len(set(candidate_names))):
            county_names_full.append(county)

    data = {
        "county_name": county_names_full,
        "candidate_name": candidate_names,
        "election_day": election_day,
        "absentee_mail": absentee_mail,
        "early_in_person": early_in_person,
        "provisional": provisional,
        "total": total
    }

    print(election_name + " scraped!")

    votes_data = pd.DataFrame(data)
    votes_data.to_csv("data/RAW_" + election_name + "_General Election.csv", index=False)

```

```python
link_2016 = "https://results.sos.ga.gov/results/public/Georgia/elections/2016NovGen"
link_2018 = "https://results.sos.ga.gov/results/public/Georgia/elections/2018NovGen"
link_2020 = "https://results.sos.ga.gov/results/public/Georgia/elections/2020NovGen"
link_2022 = "https://results.sos.ga.gov/results/public/Georgia/elections/2022NovGen"
link_2024 = "https://results.sos.ga.gov/results/public/Georgia/elections/2024NovGen"

scrape(link_2016)
scrape(link_2018)
scrape(link_2020)
scrape(link_2022)
scrape(link_2024)
```
