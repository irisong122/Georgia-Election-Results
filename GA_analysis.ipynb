```python
os.chdir("C:/Users/iriso/OneDrive/Desktop/Projects/Georgia")
os.getcwd()

ga_2016 = pd.read_csv("data/RAW_November 8, 2016_General Election.csv")
ga_2020 = pd.read_csv("data/RAW_November 3, 2020_General Election.csv")
ga_2024 = pd.read_csv("data/RAW_November 5, 2024_General Election.csv")

ga_2016["date"] = "November 8, 2016"
ga_2016["year"] = 2016

ga_2020["date"] = "November 3, 2020"
ga_2020["year"] = 2020

ga_2024["date"] = "November 5, 2024"
ga_2024["year"] = 2024

data_pres = pd.concat([ga_2016, ga_2020, ga_2024], ignore_index = True)

# cleaning

cols = ["election_day", "absentee_mail", "early_in_person", "provisional", "total"]

for col in cols:
    data_pres[col] = data_pres[col].astype(str).str.replace(',', '').astype(int)

# calculations

#print(data_pres)

data_pres["county_total"] = data_pres.groupby(["year", "county_name"])["total"].transform("sum")
data_pres["count_total_perc"] = data_pres["total"] / data_pres["county_total"] * 100

party_data = data_pres[data_pres["candidate_name"].str.contains("REP|DEM")]
party_data["party"] = party_data["candidate_name"].str.extract(r'\((.*?)\)')

# party_data = party_data.pivot(
#     index = ["county_name", "year"],
#     columns = "party",
#     values = ["county_total"])

party_data.columns = party_data.columns.map('_'.join)


print(party_data)

# export

#party_data.to_csv("webpage/data/party_data.csv")
```
